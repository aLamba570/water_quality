Suspended Matter 

```python
d11 = pd.read_csv('spmg.csv')
d11.head()


fig11 = px.scatter(d11, x = 'Dates' , y = 'Suspended Matter' ,  color='Suspended Matter', 
              labels = {
                  'x': 'Time',
                  'y': 'suspended matter',
              },title = 'Data Representation for Dissolved Organic Matter in Nawab Munshi Hussain Khan Lake')

d1_ = pd.read_csv('spmg1.csv')
d1_.head()


fig1_ = px.scatter(d1_, x = 'Dates' , y = 'Suspended Matter' ,  color='Suspended Matter', 
              labels = {
                  'x': 'Time',
                  'y': 'suspended matter',
              },title = 'Data Representation for Dissolved Organic Matter in Nawab Munshi Hussain Khan Lake')

d2_ = pd.read_csv('spmg2.csv')
d2_.head()


fig2_ = px.scatter(d2_, x = 'Dates' , y = 'Suspended Matter' ,  color='Suspended Matter', 
              labels = {
                  'x': 'Time',
                  'y': 'suspended matter',
              },title = 'Data Representation for Dissolved Organic Matter in Nawab Munshi Hussain Khan Lake')

d3_ = pd.read_csv('spmg3.csv')
d3_.head()


fig3_ = px.scatter(d3_, x = 'Dates' , y = 'Suspended Matter' ,  color='Suspended Matter', 
              labels = {
                  'x': 'Time',
                  'y': 'suspended matter',
              },title = 'Data Representation for Dissolved Organic Matter in Nawab Munshi Hussain Khan Lake')

d4_ = pd.read_csv('spmg4.csv')
d4_.head()


fig4_ = px.scatter(d4_, x = 'Dates' , y = 'Suspended Matter' ,  color='Suspended Matter', 
              labels = {
                  'x': 'Time',
                  'y': 'suspended matter',
              },title = 'Data Representation for Dissolved Organic Matter in Nawab Munshi Hussain Khan Lake')

fig_subplots = make_subplots(rows=5, cols=1)

for trace in fig11.data:
    fig_subplots.add_trace(
        trace,
        row=1, col=1
    )
for trace in fig1_.data:
    fig_subplots.add_trace(
        trace,
        row=2, col=1
    )
for trace in fig2_.data:
    fig_subplots.add_trace(
        trace,
        row=3, col=1
    )
for trace in fig3_.data:
    fig_subplots.add_trace(
        trace,
        row=4, col=1
    )
for trace in fig4_.data:
    fig_subplots.add_trace(
        trace,
        row=5, col=1
    )

## x and y axies in fig_subplots["layout"] are called xaxis, xaxis2, ..., yaxis, yaxis2, ...
## here we are making the assumption you are stacking your plots vertically
def modify_axis_titles(fig_subplots, px_fig, nrow):
    xaxis_name, yaxis_name = f"xaxis{nrow}", f"yaxis{nrow}"
    fig_subplots['layout'][xaxis_name]['title'] = px_fig.layout['xaxis']['title']
    fig_subplots['layout'][yaxis_name]['title'] = px_fig.layout['yaxis']['title']
        
for px_fig, nrow in zip([fig11, fig1_, fig2_ ,fig3_ ,fig4_],[1,2,3,4]):
    modify_axis_titles(fig_subplots, px_fig, nrow)

fig_subplots.show()

```
