```
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.graph_objs as go
import matplotlib.pyplot as plt
```

# **Analysis of Data**

## Histplots with zone divisions.

### Suspended sediment concentration

```
limit = 0.600
fig, ax = plt.subplots(figsize=(20, 8))
sns.histplot(x='Suspended sediment concentration', data=df[df['Suspended sediment concentration'] <= limit], color='green', label=f'Suspended sediment concentration <= {limit}')
sns.histplot(x='Suspended sediment concentration', data=df[df['Suspended sediment concentration'] > limit], color='yellow', label=f'Suspended sediment concentration > {limit}', ax=ax)
ax.set_xlabel('Suspended sediment concentration')
ax.set_ylabel('Count')
ax.legend()
plt.show()
```

>**100%** of the **Suspended Sediment Concentration** is in the **Need treatment zone**

### Temperature

```
fig, ax = plt.subplots(figsize=(20, 8))
sns.histplot(x='Temperature', data=df[(df['Temperature'] <= 35) & (df['Temperature'] >= 15)], color='green', label='15 <= Temperature <= 35')
sns.histplot(x='Temperature', data=df[((df['Temperature'] > 35) & (df['Temperature'] < 40)) | ((df['Temperature'] < 15)&(df['Temperature'] > 4))], color='yellow', label='4 < Temperature < 15 or 35 < Temperature < 40', bins = np.arange(4, 40))
sns.histplot(x='Temperature', data=df[(df['Temperature'] >= 40) | (df['Temperature'] <= 4)], color='red', label='Temperature >= 40 or Temperature <= 4', bins = np.arange(0, 100))
ax.set_xlabel('Temperature')
ax.set_ylabel('Count')
ax.legend()
plt.show()
```



> 1. **Temperature** has **equal contribution in all the zones**
2. Values **<=0 and >=45** are **outliers**



### Chlorophyll

```
fig, ax = plt.subplots(figsize=(20, 8))
sns.histplot(x='Chlorophyll', data=df[(df['Chlorophyll'] <= 0.1) & (df['Chlorophyll'] >= -0.1)], color='green', label='Chlorophyll <= 0.1')
sns.histplot(x='Chlorophyll', data=df[(df['Chlorophyll'] > 0.1) & (df['Chlorophyll'] < 0.5)], color='yellow', label='0.1 < Chlorophyll < 0.5')
sns.histplot(x='Chlorophyll', data=df[(df['Chlorophyll'] >= 0.5)], color='red', label='Chlorophyll >= 0.5',bins=np.arange(0.5, 1, 0.01))
ax.set_xlabel('Chlorophyll')
ax.set_ylabel('Count')
ax.legend()
plt.show()
```

# **Individual Plotting**

## Correlations

```
plt.figure(figsize=(25,6))
sns.heatmap(df1.corr(),annot=True,cmap=sns.diverging_palette(230, 20, as_cmap=True),vmax=.3, center=0,square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```
dfCorr = df1.corr()
filteredDf = dfCorr[((dfCorr >= .5) | (dfCorr <= -.5)) & (dfCorr !=1.000)]
plt.figure(figsize=(8,5))
sns.heatmap(filteredDf, annot=True, cmap=sns.diverging_palette(230, 20, as_cmap=True))
plt.show()
```

>1.   **MNDWI and Suspened sediment concentration** have a **strong positive correlation of 0.97**. MNDWI is more sensitive to water features mixed with vegetation as the NIR band is not used in its calculation. 

>2.   **Turbidity and Dissolved oxygen** have a **strong positive correlation of 0.93**. This correlation is a bit unexplainableas of now. Everyone has noticed that the more negative the turbidity, the less the water is turbid. We need more in depth analysis of this property to understand this correlation trend.

>3. **Turbidity and NDSI** have a **strong positive correlation of 0.91**. Typically, turbidity and NDSI should have a negative correlation as the temperature increases when turbidity increases. This will reduce the NDSI. But here, the opposite is going on, i.e. with the increase of turbidity, NDSI is also increasing which should not happen.

>4. **Dissolved oxygen and NDSI** have a **strong positive correlation of 0.89**. This is because cold water holds more oxygen as compared to warm water. Therefore, both the values go up and down together, hence te positive correlation.

>5. **MNDWI and pH** have a **strong negative correlation of -0.81**. This means that the pH is decreasing as the MNDWI is increasing, which should not happen because the pH has to stay close to neutral.

```
df_new = df[(df['Temperature'] > 0) | (np.isnan(df['Temperature']))]
df_new = df_new[((df_new['Dissolved oxygen'] <= 10) & (df_new['Dissolved oxygen'] >= 2)) | (np.isnan(df_new['Dissolved oxygen']))]
df_new
```

```
df4=df_new.filter(['Suspended sediment concentration', 'Dissolved Organic Matter', 'Temperature','Chlorophyll', 'Turbidity', 'Salinity', 'Dissolved oxygen', 'pH'])
```

```
import plotly.express as px
import matplotlib.pyplot as plt
for i in df4.columns:
  fig = px.box(df_new, y=i)
  plt.figsize=(18,4)
  fig.show("png")
```

```
df5 = pd.DataFrame()
for i in df4.columns:
    r_mean_daily_max = df_new.groupby(['Year', 'Month','Day'], as_index=False)[i].max()
    mda8 = r_mean_daily_max.groupby(['Year', 'Month'], as_index=False)[i].median()
    df5[i]=r_mean_daily_max.groupby(['Year', 'Month'], as_index=True)[i].median()
```
